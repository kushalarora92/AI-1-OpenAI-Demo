```python
import os
import openai

from dotenv import load_dotenv

load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY_2")

openai.api_key = OPENAI_API_KEY
```

```python
import pandas as pd
pd.DataFrame(openai.Model.list()['data'])
```

```python
## Chat completion

prompt = "Hello, how are you?"
response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": prompt}
    ],
)


print(response['choices'][0]['message']['content'])
```

```python
## Chat completion with multiple prompts

prompt1 = "Hello, how are you?"
prompt2 = "I am a 25 years old & I am a programmer"
prompt3 = "Tell me about me"

response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": prompt1},
        {"role": "user", "content": prompt2},
        {"role": "user", "content": prompt3}
    ],
)

print(response['choices'][0]['message']['content'])
```

```python
# Parameters max_token

response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=[{"role": "user", "content": prompt}],
    max_tokens=25
)

print(response['choices'][0]['message']['content'])
```

```python
# Parameters temperature

response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=[{"role": "user", "content": prompt}],
    temperature=0.6,
    n=2 # number of responses from the model
)

print(response['choices'][0]['message']['content'])
```

```python
functions = [{
    "name": "get_current_weather",
    "description": "Get the current weather in a given location",
    "parameters": {
        "type": "object",
        "properties": {
            "location": {"type": "string", "description": "The city and state, e.g. San Francisco, CA"}
        }
    },
    "required": ["location"]
}]

messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "What is the weather in San Francisco?"}
    ]
response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=messages,
    functions=functions,
    function_call="auto"
)
```
